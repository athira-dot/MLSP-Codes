```python
#%%
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.decomposition import PCA
```

```python

data_path='Data/emotion_classification/train/'
files=os.listdir(data_path)
test_data_path='Data/emotion_classification/test/'
```

```python
#os.listdir(test_data_path)
#files=os.listdir(data_path)
test_files=os.listdir(test_data_path)
#print(os.listdir(data_path))

N=len(files)
label=np.zeros(N)
j=0
for i in files:
  w=i.split(".")
  if (w[1]=="happy"):
    label[j]=1
  else:
    label[j]=-1
  j=j+1
label=np.array(label)

n=len(test_files)
label_test=np.zeros(n)
j=0
for i in test_files:
  w=i.split(".")
  if (w[1]=="happy"):
    label_test[j]=1
  else:
    label_test[j]=-1
  j=j+1
label_test=np.array(label_test)



```

```python
train_imag=[]
test_imag=[]
for i in files:
  ind_path=data_path+'/'+i
  im=Image.open(ind_path)
  #plt.imshow(im)
  #plt.show()
  im=np.array(im).reshape(-1,)
  train_imag.append(im)

X=train_imag   #data

for i in test_files:
  ind_path=test_data_path+'/'+i
  img=Image.open(ind_path)
  #plt.imshow(im)
  #plt.show()
  img=np.array(img).reshape(-1,)
  test_imag.append(img)

Y=test_imag   #data
```

```python
comb_data=np.vstack([X,Y])
print(np.shape(comb_data))

#calculating the mean of data
#X=np.array(X)
#mean=np.mean(X,axis=0)
#X_new=X-mean

#Y=np.array(Y)
#mean_Y=np.mean(Y,axis=0)
#Y_new=Y-mean_Y

comb_data=np.array(comb_data)
mean_comb=np.mean(comb_data,axis=0)
comb_new=comb_data-mean_comb
```

```python

#%%
#applying high dimensional PCA on data 
def pca_(K,comb_new):
#K=10
    pca = PCA(n_components=K)
#pca_x = pca.fit_transform(X) #testing
#pca_y = pca.fit_transform(Y)  #training

    pca_comb=pca.fit_transform(comb_new)

    pca_train=pca_comb[0:20,:]   #test data 
    pca_test=pca_comb[20:,:]     #train data
    
    return(pca_train,pca_test)

```

```python
stat=[]
def svm_(label,pca_train,ker,pca_test,label_test):
    p=svm_problem(label,pca_train)

    param = svm_parameter()

#RBF kernel
    c=0.000001
    while(c<=100000): 

        e=0.1
        while(e<=1):

            param.kernel_type=ker
            param.C=c
            param.eps=e

  ## training  the model
            model_=svm_train(p,param)


        
            #print("for training data")
            PREDICT_TEST=svm_predict(label,pca_train,model_)

            #print("for testing data")
            PREDICT_TRAIN=svm_predict(label_test,pca_test,model_)
            
            #print([K,c,e,ker,PREDICT_TEST[1][0],PREDICT_TRAIN[1][0]])
            stat.append([K,c,e,ker,PREDICT_TEST[1][0],PREDICT_TRAIN[1][0]])
            e=e+0.5
        c=c*10

    
```

```python
#for training printing the K,C,epsilon,kernel_type  with maximum training accuracy
print("training output")
for i in range(40):
    low=24*i
    up=24*(i+1)

    max_index=low
    max_value=stat[low][4]
    for i in range(low,up):
            if(stat[i][4]>max_value):
                max_index=i
                max_value=stat[i][4]
    print(stat[max_index])
    
#here stat[i][0]=K
#stat[i][1]=c
#stat[i][2]=epsilon
#stat[i][3]=kernel type , 0-linear, 1-poly ,2-RBF ,3-SIGMOID
#stat[i][4]=training accurcy
#stat[i][5]=testing accurcy 
```
